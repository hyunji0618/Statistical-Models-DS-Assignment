## Generalized linear models

#### 1. For each of the situations below, identify the response and explanatory variables, variable types, and the generalized linear model that is well-suited to model the data. Make sure to justify your answer. Write down the linear predictor as well as the link function.

##### **Scenario 1**: The effect of age, sex, height, daily food intake and minutes of daily exercise on a person's weight.
- **Response Variable**: Weight (Continuous)
- **Explanatory Variables**: Age (Continuous), Sex (Categorical), Height (Continuous), Daily Food Intake (Continuous), Minutes of Daily Exercise (Continuous)
- **GLM Model**: Linear regression model (Weight is a continuous variable, and the response variable can be assumed to have a linear relationship with the explanatory variables)
- **Linear Predictor**: $ g(\mu_i) = \mathbf{x}_i^T \boldsymbol{\beta} = \beta_0 + \beta_1 \cdot \text{Age} + \beta_2 \cdot \text{Sex} + \beta_3 \cdot \text{Height} + \beta_4 \cdot \text{Daily Food Intake} + \beta_5 \cdot \text{Minutes of Daily Exercise} $
- **Link Function**: $g(\mu)=\mu$

##### **Scenario 2**: The effect of sex, age, GPA, major, prior years of work experience, and prior income levels on whether a full-time graduate students finds employment upon graduation
- **Response Variable**: Employment Status (Binary: Employed / Unemployed)
- **Explanatory Variables**: Age (Continuous), Sex (Categorical), GPA (Continuous), Major (Categorical), Prior Years of Work Experience (Continuous), Prior Income Levels (Categorical)
- **GLM Model**: Logistic regression model (Employment Status is a binary, categorical variable)
- **Linear Predictor**: $ g(\mu_i) = \mathbf{x}_i^T \boldsymbol{\beta} = \beta_0 + \beta_1 \cdot \text{Age} + \beta_2 \cdot \text{Sex} + \beta_3 \cdot \text{GPA} + \beta_4 \cdot \text{Major} + \beta_5 \cdot \text{Prior Years of Work Experience} + \beta_6 \cdot \text{Prior Income Levels}$
- **Link Function**: $g(\mu) = \log \left(\frac{\mu}{1-\mu}\right)$

##### **Scenario 3**: The number of mortgage loan defaults in a given year by different counties across the United States. For each household/borrower information on income, loan interest rate, age, debt, loan to value at origination are available.
- **Response Variable**: Number of loan defaults (Count data)
- **Explanatory Variables**: Income (Continuous), Loan Interest Rate (Continuous), Age (Continuous), Debt (Continuous), Loan to Value at Organization (Continuous)
- **GLM Model**: Poisson regression model (The number of loan defaults is a count data)
- **Linear Predictor**: $ g(\mu_i) = \mathbf{x}_i^T \boldsymbol{\beta} = \beta_0 + \beta_1 \cdot \text{Income} + \beta_2 \cdot \text{Loan Interest Rate} + \beta_3 \cdot \text{Age} + \beta_4 \cdot \text{Debt} + \beta_5 \cdot \text{Loan to Value at Organization}$
- **Link Function**: $g(\mu)=log(\mu)$

### 2. Assume that the linear predictor, $\mathbf{x}_i^T \boldsymbol{\beta}$, takes the values between âˆ’3 and 3 equally spaced by 0.01. If the link function is the natural logarithm, $log(\mu_i) = \mathbf{x}_i^T \boldsymbol{\beta}$, then compute and plot the mean response, $\mu_i$, against the linear predictor, $\mathbf{x}_i^T \boldsymbol{\beta}$.

```python
import numpy as np
import matplotlib.pyplot as plt

lin_predictor1 = np.arange(-3, 3.01, 0.01)
mu1 = np.exp(lin_predictor1)

plt.figure(figsize=(7, 5))
plt.plot(lin_predictor1, mu1)
plt.xlabel(r"Linear Predictor $(\mathbf{x}_i^T \boldsymbol{\beta})$")
plt.ylabel(r"Mean Response ($\mu$)")
plt.title(r"Mean Response vs Linear Predictor (Link Function: $log(\mu_i) = \mathbf{x}_i^T \boldsymbol{\beta}$)", fontsize=12)
plt.grid(True)
plt.show()
```

#### 3. Assume that the linear predictor, $\mathbf{x}_i^T \boldsymbol{\beta}$, takes the values between âˆ’5 and 5 equally spaced by 0.01. If the link function is the logit, $\log \left(\frac{\mu_i}{1-\mu_i}\right) =  \mathbf{x}_i^T \boldsymbol{\beta}$, then compute and plot the mean response, $\mu_i$, against the linear predictor, $\mathbf{x}_i^T \boldsymbol{\beta}$.

```python
lin_predictor2 = np.arange(-5, 5.01, 0.01)
mu2 = np.exp(lin_predictor2) / (1 + np.exp(lin_predictor2))

plt.figure(figsize=(7, 5))
plt.plot(lin_predictor2, mu2)
plt.xlabel(r"Linear Predictor $(\mathbf{x}_i^T \boldsymbol{\beta})$")
plt.ylabel(r"Mean Response ($\mu$)")
plt.title(r"Mean Response vs Linear Predictor (Link Function: $\log \left(\frac{\mu_i}{1-\mu_i}\right) =  \mathbf{x}_i^T \boldsymbol{\beta}$)", fontsize=12)
plt.grid(True)
plt.show()
```

#### 4. Assume that the linear predictor, $\mathbf{x}_i^T \boldsymbol{\beta}$, takes the values between 0.5 and 5 equally spaced by 0.01. If the link function is negative inverse, $ -\mu_i^{-1}= \mathbf{x}_i^T \boldsymbol{\beta}$, then compute and plot the mean response, $\mu_i$, against the linear predictor, $\mathbf{x}_i^T \boldsymbol{\beta}$.

```python
lin_predictor3 = np.arange(0.5, 5.01, 0.01)
mu3 = -1 / lin_predictor3

plt.figure(figsize=(7, 5))
plt.plot(lin_predictor3, mu3)
plt.xlabel(r"Linear Predictor $(\mathbf{x}_i^T \boldsymbol{\beta})$")
plt.ylabel(r"Mean Response ($\mu$)")
plt.title(r"Mean Response vs Linear Predictor (Link Function: $ -\mu_i^{-1}= \mathbf{x}_i^T \boldsymbol{\beta}$)", fontsize=12)
plt.grid(True)
plt.show()
```
